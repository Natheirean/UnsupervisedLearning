```python
#Dependancies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
#Load Data from CSV
file = "crypto_data.csv"
crypto_df = pd.read_csv(file, index_col=0)
crypto_df
```

```python
#Drop all Crypto not being traded
crypto_active = crypto_df.loc[crypto_df.IsTrading==True].drop(columns="IsTrading")
crypto_active
```

```python
#Remove rows with N/A
crypto_active = crypto_active.dropna(axis=0)
crypto_active
```

```python
#Filter for coins that have been mined
crypto_active = crypto_active.loc[crypto_active.TotalCoinsMined > 0]
crypto_active
```

```python
#Remove CoinName Column
crypto_active = crypto_active.drop(columns="CoinName")
crypto_active
```

```python
crypto_dumb = pd.get_dummies(crypto_active[['Algorithm', 'ProofType']])
crypto_dumb = crypto_dumb.merge(crypto_active[['TotalCoinsMined', 'TotalCoinSupply']], how='inner', left_index=True, right_index=True)
crypto_dumb
```

```python
#Standardize Data
crypto_scaled = StandardScaler().fit_transform(crypto_dumb)
```

```python
pca = PCA(n_components=.9)
df_pca = pca.fit_transform(crypto_scaled)
df_pca
```

```python
tsne = TSNE(learning_rate=100)
tsne_features = tsne.fit_transform(df_pca)
tsne_df = pd.DataFrame(data=tsne_features)
tsne_df
```

```python
plt.scatter(tsne_df[0], tsne_df[1])
plt.show()
```

```python
inertia = []
k=list(range(1,10))

for i in k:
    km = KMeans(n_clusters=i)
    km.fit(tsne_df)
    inertia.append(km.inertia_)
    
elbow_data ={"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,10))
plt.xlabel("Clusters")
plt.ylabel("Intertia")
plt.title("Elbow Curve")
plt.show()
```

```python
#Crypto can be clustered together, there is a distict elbow between 3 and 4 clusters,
#so we can conclude that cyrpto can be clustered into 3 or 4 clusters.
```
